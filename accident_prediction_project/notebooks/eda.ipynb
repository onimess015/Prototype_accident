# SafeRoute AI EDA
Exploratory data analysis for the Road Accident Risk Prediction System.

## Objectives
- Inspect the main training dataset
- Review missing values and target distribution
- Visualize feature distributions
- Review numeric correlations

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.data_loader import load_main_dataset
from src.feature_engineering import select_features_and_target
from src.preprocessing import clean_dataframe

plt.style.use('ggplot')
```

```python
df = clean_dataframe(load_main_dataset())
df.shape, df.head()
```

```python
missing_summary = df.isna().sum().sort_values(ascending=False)
missing_summary[missing_summary > 0].head(20)
```

```python
X, y = select_features_and_target(df)
severity_distribution = y.value_counts().rename(index={0: 'Low/Minor', 1: 'Severe/Fatal'})
severity_distribution
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pd.Series(y).value_counts().sort_index().plot(kind='bar', ax=axes[0], title='Binary Target Distribution')
axes[0].set_xticklabels(['Low/Minor', 'Severe/Fatal'], rotation=0)
X['driver_age'].plot(kind='hist', bins=20, ax=axes[1], title='Driver Age Distribution')
plt.tight_layout()
```

```python
categorical_columns = [column for column in X.columns if X[column].dtype == 'string' or X[column].dtype == object]
categorical_columns
```

```python
top_vehicle_types = X['vehicle_type'].astype('string').value_counts().head(10)
top_vehicle_types.plot(kind='bar', figsize=(10, 4), title='Top Vehicle Types Involved')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
numeric_frame = X.select_dtypes(include=['number']).copy()
correlation = numeric_frame.corr(numeric_only=True)
correlation
```

## Observations
- The dataset provides strong coverage for driver, vehicle, road, and environment features.
- Accident severity can be reduced to a clean binary target for risk modeling.
- Numeric features such as driver age and speed limit are suitable for scaling.
- Most operational features are categorical and benefit from one-hot encoding for classical ML models.
